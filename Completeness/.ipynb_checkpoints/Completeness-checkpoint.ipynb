## Load packages/libraries

```python
import pandas as pd
import numpy as np
import os
import random
```

## Setup

```python
cwd = os.getcwd()
pathData = os.path.abspath(os.path.join(cwd,"..","00. Data"))
```

## Load dataset

```python
raw_df = pd.read_excel(os.path.join(pathData, "loans_mds.xlsx"), sheet_name = "Sheet1", engine = "openpyxl")
factors = raw_df.columns.to_list()
raw_df.shape
```

```python
year_list = []
for i in range(len(raw_df)):
    year_list.append(random.random())

raw_df['YEAR_temp'] = year_list
raw_df['YEAR'] = np.where(raw_df['YEAR_temp'] <= 0.2, 2017,
                         np.where(raw_df['YEAR_temp'] <= 0.4, 2018,
                                 np.where(raw_df['YEAR_temp'] <= 0.6, 2019,
                                         np.where(raw_df['YEAR_temp'] <= 0.8, 2020, 2021))))

raw_df = raw_df.drop(columns="YEAR_temp")
```

```python
completeness_by_year = pd.DataFrame({'Variable':factors, '2017':np.nan, '2018':np.nan, '2019':np.nan, '2020':np.nan, '2021':np.nan})

for factor in factors:
    for year in range(2017,2022):
        completeness_by_year.loc[completeness_by_year['Variable'] == factor, str(year)] = len(raw_df[factor])/len(raw_df)
        
completeness_by_year
```
